```python
# models and imports
from pathlib import Path
import pandas as pd
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_ollama import ChatOllama, OllamaEmbeddings

llm = ChatOllama(model="qwen:latest", temperature=0)
embedding_model = OllamaEmbeddings(model="nomic-embed-text:latest")

print("Ready")
```

```python
# config dataset and retriever
DATA_DIR = "./data"
pdf_files = sorted(Path(DATA_DIR).glob("*.pdf"))

print(f"PDFs Found: {len(pdf_files)}")
[f.name for f in pdf_files]
```

```python
## load pdf
loader = PyPDFDirectoryLoader(DATA_DIR)
documents = loader.load()

print(f"Pages Loaded: {len(documents)}")
```

```python
# insoect data
doc = documents[0]

print(doc.metadata)
print(doc.page_content[:1000])
```

```python
# creating metadata table
metadata_df = pd.DataFrame([
{
"source": Path(d.metadata["source"]).name,
"page": d.metadata.get("page",0),
"chars": len(d.page_content),
"words": len(d.page_content.split()),
"text": d.page_content
}
for d in documents
])

metadata_df.head()
```

```python
metadata_df.shape
```

```python
metadata_df.describe()
```

```python
# classify the doucments
def classify(x):
    x=x.lower()
    if "guide" in x: return "guide"
    if "form" in x or "declaration" in x: return "form"
    if "tax" in x: return "tax"
    if "policy" in x: return "policy"
    if "faq" in x: return "faq"
    return "other"

metadata_df["doc_type"] = metadata_df["source"].apply(classify)

metadata_df[["source","doc_type"]].head()
```
